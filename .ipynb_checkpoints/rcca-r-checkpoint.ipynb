```r
library("pacman")

pacman::p_load("mixOmics", "tidyverse")
```

```r
data(nutrimouse)
X <- nutrimouse$lipid
print(dim(X))

Y <- nutrimouse$gene
```

```r
# Estimate lambda1 and lambda2
grid1 <- seq(0, 0.2, length = 5) 
grid2 <- seq(0.0001, 0.2, length = 5)
cv <- tune.rcc(X, Y, grid1 = grid1, grid2 = grid2, validation = "loo")
par(mfrow=c(1,1))

# proc rcca
nutrimouse.rcc <- rcc(X,Y, ncomp = 3, lambda1 = cv$opt.lambda1,lambda2 = cv$opt.lambda2)
```

```r
png(filename="rcca-nutrimouse-barplot.png")
plot(nutrimouse.rcc, scree.type = "barplot")
dev.off()
```
